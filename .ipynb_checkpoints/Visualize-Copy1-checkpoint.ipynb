```python
results_dir = '../../../../media/lndunn/datadrive/projects/outagesim/simulations/'

metadata = pd.read_csv(os.path.join(results_dir, 'metadata.csv'), index_col=0, parse_dates=['timestamp',])
metadata = metadata.sort_values(['number','timestamp'])
metrics = pd.DataFrame(index=metadata.index, columns=['saidi','saifi','caidi', 'kwh'])
for i, row in metadata.iterrows():
    outages = pd.read_csv(os.path.join(results_dir, 'simulation%i.csv'%(row['number'])))
    
    for key in outages.keys():
        idx = pd.isnull(outages[key])
        outages[key].loc[idx] = 0
        
    if sum(outages['customers'])==0:
        continue
#     if len(outages) == 0 or sum(outages['customers']==0) or (sum(pd.notnull(outages['customers']))==0):
#         continue
        
    idx = pd.notnull(outages['customers'])
    metrics.loc[i] = [(outages['customers']*(outages['end']-outages['start'])).sum()/system.bus_outages['customers'].sum(),
                        outages['customers'].sum()/system.bus_outages['customers'].sum(),
                        np.average((outages['end']-outages['start'])[idx], weights=outages['customers'][idx]),
                        (outages['load']*(outages['end']-outages['start'])).sum()/60./(system.bus_outages['load'].sum()*(outages['end'].max()-outages['start'].min())),
                        ]
    
rates = metadata[['overhead_outage_rate','underground_outage_rate']].drop_duplicates()
for i, row in rates.iterrows():
    idx = metadata['overhead_outage_rate'] == row['overhead_outage_rate']
    idx &= metadata['underground_outage_rate'] == row['underground_outage_rate']
#     idx &= pd.isnull(metrics).sum(axis=1)==0
#     print row, sum(idx)
#     print metrics[idx]
    fig, ax = plt.subplots(figsize=(15,2.5), ncols=len(metrics.keys()), nrows=1)
    for i, key in enumerate(metrics.keys()):
        _idx = idx & pd.notnull(metrics[key])
        if key in ['saidi','saifi']:
            ax[i].hist((365.*metrics[key][_idx]).tolist(), bins=20)
        else:
            ax[i].hist(metrics[key][_idx].tolist(), bins=20)
        ax[i].set_xlabel(key, fontsize=14)
    plt.show()
```

```python
class GridNetwork(object):
    def __init__(self, path_to_network):
        print 'ingesting network info'
        self.network = pp.from_json(path_to_network)
        self.bus_index = self.bus_names_to_numbers()
        self.loads = self.map_loads_to_busses()
        
        print 'calculating outage implications'
        groups = self.loads.groupby('bus')
        self.bus_outages = pd.DataFrame(columns=['customers','load','cost'], index=groups.groups.keys())
        self.bus_outages['customers'] = groups['zone'].count()
        self.bus_outages['load'] = groups['kW'].sum()
        
#         self.affected_busses = self.get_affected_busses()
        
#         if 'line_outage_consequences.csv' in os.listdir(os.getcwd()):
#             self.line_outages = pd.read_csv('line_outage_consequences.csv', index_col='line')
#         else:
#             self.line_outages = pd.DataFrame(columns=['customers','load','cost'], index=self.network['line'].index)

#         it = 0
#         idx = pd.isnull(self.line_outages['customers'])
#         lines_to_process = self.line_outages[idx].index
#         for l, bus_list in self.affected_busses.loc[lines_to_process].iteritems():
#             if pd.isnull(bus_list):
#                 pass
#             else:
#                 bus_list = np.array(bus_list.split('-')).astype(int)
#                 bus_list = set(bus_list) & set(self.bus_outages.index)
                
#                 self.line_outages.loc[l] = self.bus_outages.loc[bus_list].sum()
#             it += 1
#             if it % 1000 == 0:
#                 print it, 'of', len(self.affected_busses)
                
#         self.line_outages.to_csv('line_outage_consequences.csv', index_label='line')

    def bus_names_to_numbers(self):
        bus_index = self.network['bus'][['name','zone']].reset_index()
        bus_index = bus_index.rename(columns={'name':'bus_name', 'index':'bus'})
        bus_index = bus_index.set_index(['zone','bus_name'])
        return bus_index
    
    def map_loads_to_busses(self):
        loads = utils.get_loads()
        loads = loads.rename(columns={'bus': 'bus_name'})
        loads = loads.join(self.bus_index, on=['zone','bus_name'])
        return loads
    
    def get_affected_busses(self):
        filenames = ['affected_busses.csv', 'affected_busses2.csv']
        affected_busses = pd.concat([pd.read_csv(f, header=None) for f in filenames])
        affected_busses = affected_busses.drop_duplicates().set_index(0)
        return affected_busses[1]
```
